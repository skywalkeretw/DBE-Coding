```python
from pandas_datareader import data, wb
import numpy as np
import pandas as pd
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
bank_stocks = pd.read_pickle('all_banks')
```

```python
bank_stocks.head()
```

```python
BAC = bank_stocks['BAC']
C = bank_stocks['C']
GS = bank_stocks['GS']
JPM = bank_stocks['JPM']
MS = bank_stocks['MS']
WFC = bank_stocks['WFC']
```

```python
BAC.head()
```

```python
tickers = ['BAC','C','GS','JPM','MS','WFC']
```

```python
pd.concat([BAC,C,GS,JPM,MS,WFC], axis=1, keys=tickers)
```

```python
for tick in tickers:
    print(tick, bank_stocks[tick]['Close'].max())
```

```python
bank_stocks.xs(key='Close', axis=1, level='Stock Info').max()
```

```python
returns = pd.DataFrame()
```

```python
for tick in tickers:
    returns[tick+ ' Return'] = bank_stocks[tick]['Close'].pct_change()
returns.head()
```
