```python
import seaborn as sns
import numpy as np
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.barplot(x="sex", y="total_bill", data=tips, estimator=np.std)
```

```python
sns.countplot(x="sex", data=tips)
```

```python
sns.boxplot(data=tips, palette="rainbow", orient="h")
```

```python
sns.boxplot(x="day", y="total_bill", hue="smoker", data=tips, palette="rainbow")
```

```python
sns.violinplot(x="day", y="total_bill", hue="sex", split=True, data=tips)
```

```python
sns.stripplot(x="day", y="total_bill",  data=tips , hue="sex",jitter=True, palette="Set1", dodge=True)
```

```python
sns.swarmplot(x="day", y="total_bill",  data=tips, hue="sex", palette="Set1" )
```

```python
sns.violinplot(x="tip", y="day", data=tips, palette="rainbow")
sns.swarmplot(x="tip", y="day", data=tips, color="black", size=3)
```

```python
# sns.factorplot(x="sex", y="total_bill", data=tips, kind="bar")
sns.catplot(x="sex", y="total_bill", data=tips, kind="bar")
```
