```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_title("Plot")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.plot(x,y)
fig.show()
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])
ax1.plot(x,y)

ax2.plot(x,y)
fig.show()
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax1.plot(x,z)
ax1.set_xlabel('X')
ax1.set_ylabel('Y')

ax2 = fig.add_axes([0.2,0.5,.4,.4])
ax2.plot(x,y)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('Zoom')
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)
fig.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y, color="blue", lw=3, ls='--')
axes[1].plot(x, z, color="red", lw=3, ls='-')
fig.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12,3))
axes[0].plot(x, y, color="blue", lw=3, ls='--')
axes[1].plot(x, z, color="red", lw=3, ls='-')
fig.show()
```
