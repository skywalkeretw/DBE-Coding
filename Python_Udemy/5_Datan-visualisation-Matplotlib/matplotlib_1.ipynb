```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 5, 11)
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x, y, 'r')
plt.title("plot")
plt.xlabel("X Achse")
plt.ylabel("Y Achse")
plt.show()
```

```python
plt.subplot(1,2, 1)
plt.plot(x, y, 'r--')
plt.subplot(1,2, 2)
plt.plot(y, x, 'g*-')
plt.show()
```

```python
af = plt.figure()
axes = af.add_axes([0.1, 0.1,0.8,0.8])
axes.plot(x,y,'b')
axes.set_title("Title")
axes.set_xlabel("X Achse")
axes.set_ylabel("Y Achse")
```

```python
af = plt.figure()
axes1 = af.add_axes([0.1, 0.1,0.8,0.8])
axes1.plot(x,y,'b')
axes1.set_title("Axes1")
axes1.set_xlabel("X Achse")
axes1.set_ylabel("Y Achse")

axes2 = af.add_axes([0.6, 0.2,0.4,0.3])
axes2.plot(x,y,'y')
axes2.set_title("Axes2")
axes2.set_xlabel("X Achse")
axes2.set_ylabel("Y Achse")

af.show()
```

```python
diag, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.plot(x,y,'g*-.')
plt.tight_layout()
```

```python
diag = plt.figure(figsize=(8,4), dpi=100)
ax = diag.add_axes([0,0,1,1])
ax.plot(x,y)
diag.show()
```

```python
diag, axes = plt.subplots(figsize=(12,3))
axes.plot(x,y)
diag.show()
```

```python
diag.savefig("myplot.png", dpi=200)
```

```python
diag = plt.figure()
ax = diag.add_axes([0,0,1,1])
ax.plot(x, x**2, label = "x**2")
ax.plot(x, x**3, label = "x**3")
ax.legend(loc=10)
diag.show()
```

```python
diag, ax = plt.subplots()
ax.plot(x, x**2, color="blue", linewidth=0.5, linestyle='--', 
        marker="o", markersize=20,)
diag.show()
```

```python
diag, ax = plt.subplots()
ax.plot(x, x**2, color="blue")
ax.set_xlim([0,1])
ax.set_ylim([0,5])
diag.show()
```
